```python
import os
import pandas as pd
from numpy import argmax
from statsmodels.tsa.api import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
```

```python
data_dir = "./data/raw_price_train"
all_data = dict()
for i in range(1, 9):
    fp = os.path.join(data_dir, "%d_r_price_train.csv" % i)
    df = pd.read_csv(fp, header=0, index_col=0, parse_dates=["Date"])
    df = df.loc[df.index >= pd.Timestamp("2014-01-02"), :]
    all_data[i] = df
```

```python
data_dir = "./data/raw_price_train"
max_lag = 30
for i, df in all_data.items():
    print("Stock No.%d" % i)
    for col in df.columns:
        if col != "Adj Close":
            seq = df[col].diff(1).dropna()
            print("=================== Auto correlation plots and stationary test for 1-order difference feature sequence: %s ===================" % col)
            fig = plt.figure(figsize=(12, 4))
            ax1 = fig.add_subplot(121)
            plot_acf(seq, ax=ax1, lags=max_lag)
            ax2 = fig.add_subplot(122)
            plot_pacf(seq, ax=ax2, lags=max_lag)
            plt.show()
            df_test = adfuller(seq)
            pval = df_test[1]
            acf_seq = acf(seq, nlags=max_lag)
            pacf_seq = pacf(seq, nlags=max_lag)
            max_acf_idx, max_pacf_idx = argmax(acf_seq), argmax(pacf_seq)
            print("Max auto-correlation coefficient: %g, with lag = %d; max partial auto-correlation coefficient: %g, with lag = %d." % 
                  (acf_seq[max_acf_idx], max_acf_idx, pacf_seq[max_pacf_idx], max_pacf_idx))
            print("Feature sequence: %s, 1-order difference ADF Test P-value: %g" % (col, pval))
```
